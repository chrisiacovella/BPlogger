```python
import BP
```

After initializing the logger, we can add data manually which will be stored in a pandas dataframe.

```python
bp_log = BP.BP_log()

bp_log.add_reading(data={'systolic': 115, 'diastolic': 80, 'heartrate': 65, 'timestamp': '2023-04-01 10:00'})
bp_log.add_reading(data={'systolic': 120, 'diastolic': 77, 'heartrate': 64, 'timestamp': '2023-04-02 10:01'})
bp_log.add_reading(data={'systolic': 110, 'diastolic': 75, 'heartrate': 61, 'timestamp': '2023-04-03 10:56'})
```

```python
bp_log.bp_readings
```

Save to a CSV file.

```python
bp_log.save('test.csv')
```

Let us read in 4 weeks of fictious data from a CSV file.

```python
bp_log = BP.BP_log()

bp_log.load('test_data.csv')
```

```python
bp_log.bp_readings
```

Display the daily summary, i.e., mean/stdev values of all readings on a given day.

```python
bp_log.daily_summary()
```

We can plot the daily summary as well:

```python
bp_log.plot_daily_summary()
```

Similarly we can display the weekly summary.

```python
bp_log.weekly_summary()
```

```python
bp_log.plot_weekly_summary()
```

We can display a histogram of the blood pressure data. 

```python
bp_log.plot_histogram()
```

We can restrict the range for the histogram. For example, considering the first half of the data

```python
bp_log.plot_histogram(end_date='2023-04-16')
```

...now consider the last half of the data.

```python
bp_log.plot_histogram(start_date='2023-04-16')
```

```python
bp_log.plot_histogram(start_date='2023-04-05', end_date='2023-04-12')
```
